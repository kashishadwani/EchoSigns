```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.regularizers import l2
# Add this import at the top of your file with the other imports
import pyttsx3

def build_improved_model(input_shape=(128, 128, 1), num_classes=27):
    weight_decay = 0.0001  # L2 regularization factor
    
    model = Sequential()
    
    # First Convolutional Block - Simplified
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu', 
                     kernel_regularizer=l2(weight_decay),
                     input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    
    # Second Convolutional Block
    model.add(Conv2D(64, (3, 3), padding='same', activation='relu',
                     kernel_regularizer=l2(weight_decay)))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.3))
    
    # Third Convolutional Block
    model.add(Conv2D(128, (3, 3), padding='same', activation='relu',
                     kernel_regularizer=l2(weight_decay)))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.4))
    
    # Flatten and Dense Layers
    model.add(Flatten())
    model.add(Dense(128, activation='relu', kernel_regularizer=l2(weight_decay)))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))  # Increased dropout
    model.add(Dense(num_classes, activation='softmax'))
    
    # Compile model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model
```

```python
# PART 2: IMPROVED TRAINING WITH DATA AUGMENTATION
def train_model_with_augmentation(train_data, train_target, validation_split=0.2, epochs=25):
    # More aggressive data augmentation
    datagen = ImageDataGenerator(
        rotation_range=15,        # Increased rotation
        width_shift_range=0.15,   # Increased shift
        height_shift_range=0.15,  # Increased shift
        zoom_range=0.15,          # Increased zoom
        shear_range=0.1,          # Added shear
        brightness_range=[0.8, 1.2],  # Added brightness variation
        horizontal_flip=False,     # Don't flip for sign language 
        validation_split=validation_split  # Increased validation set
    )
    
    # Build the model
    model = build_improved_model()
    
    # Create callbacks with more aggressive early stopping
    from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
    
    checkpoint = ModelCheckpoint(
        'sign_model_antioverfitting.keras',
        monitor='val_loss',  # Changed to monitor loss instead of accuracy
        verbose=1,
        save_best_only=True,
        mode='min'  # Minimize loss
    )
    
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=5,  # Reduced patience - stop earlier
        verbose=1,
        restore_best_weights=True
    )
    
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=2,  # Reduced patience for LR changes
        verbose=1,
        min_lr=0.00001
    )
    
    callbacks = [checkpoint, early_stopping, reduce_lr]
    
    # Train with smaller batch size
    train_generator = datagen.flow(
        train_data, train_target,
        batch_size=16,  # Reduced batch size
        subset='training'
    )
    
    validation_generator = datagen.flow(
        train_data, train_target,
        batch_size=16,  # Reduced batch size
        subset='validation'
    )
    
    # Train the model
    history = model.fit(
        train_generator,
        epochs=epochs,
        validation_data=validation_generator,
        callbacks=callbacks
    )
    
    # Save the final model
    model.save('asl_classifier_improved.h5')
    
    return model, history
```

```python
# PART 3: IMPROVED PREPROCESSING FUNCTION
def preprocess_image(img, img_size=128, min_value=70):
    """Consistent preprocessing for both training and inference"""
    # Convert to grayscale if image has 3 channels
    if len(img.shape) == 3 and img.shape[2] == 3:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        gray = img
    
    # Apply Gaussian blur
    blur = cv2.GaussianBlur(gray, (5, 5), 2)
    
    # Apply adaptive thresholding
    th3 = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)
    
    # Apply Otsu's thresholding
    ret, res = cv2.threshold(th3, min_value, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    
    # Resize to the target size
    resized = cv2.resize(res, (img_size, img_size))
    
    # Normalize
    normalized = resized / 255.0
    
    # Reshape for the model
    reshaped = np.reshape(normalized, (1, img_size, img_size, 1))
    
    return reshaped
```

```python
# PART 4: IMPROVED REAL-TIME DETECTION WITH MEDIAPIPE HAND DETECTION
import cv2
import numpy as np
import mediapipe as mp
from collections import deque, Counter
from tensorflow.keras.models import load_model

def real_time_sign_detection():
    # Load the trained model
    model = load_model("asl_classifier_improved.h5")
    
    # Initialize MediaPipe Hand detector
    mp_hands = mp.solutions.hands
    hands = mp_hands.Hands(
        static_image_mode=False,
        max_num_hands=1,
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5
    )
    mp_drawing = mp.solutions.drawing_utils
    
    # Labels dictionary
    labels_dict = {
        0: '0', 1: 'A', 2: 'B', 3: 'C', 4: 'D', 5: 'E', 6: 'F', 7: 'G', 8: 'H', 
        9: 'I', 10: 'J', 11: 'K', 12: 'L', 13: 'M', 14: 'N', 15: 'O', 16: 'P', 
        17: "Q", 18: 'R', 19: 'S', 20: 'T', 21: 'U', 22: 'V', 23: 'W', 24: 'X', 
        25: 'Y', 26: 'Z'
    }
    
    # Parameters
    img_size = 128
    min_value = 70
    confidence_threshold = 0.7
    
    # For temporal smoothing
    recent_predictions = deque(maxlen=10)
    
    # For sentence building
    sentence = ""
    prev_letter = None
    letter_counter = 0
    letter_threshold = 15  # Number of consistent frames to register a letter
    
    # For space adding
    space_counter = 0
    space_threshold = 30  # Frames with no hand to add space
    
    # Start video capture
    cap = cv2.VideoCapture(0)

    # Initialize text-to-speech engine
    tts_engine = pyttsx3.init()
    tts_engine.setProperty('rate', 150)  # Speed of speech
    tts_engine.setProperty('volume', 0.9)  # Volume (0 to 1)
    
    # Optional: Change voice (uncomment if you want to change the voice)
    # voices = tts_engine.getProperty('voices')
    # tts_engine.setProperty('voice', voices[1].id)  # Usually index 1 is female voice
    
    # Add speech queue to avoid repeated speaking of the same text
    last_spoken_text = ""
    speak_requested = False
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            break
        
        # Flip the frame horizontally for a more intuitive mirror view
        frame = cv2.flip(frame, 1)
        
        # Convert to RGB for MediaPipe
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Process with MediaPipe
        results = hands.process(rgb_frame)
        
        # Draw hand landmarks and get the hand crop if detected
        if results.multi_hand_landmarks:
            space_counter = 0  # Reset space counter when hand is detected
            
            for hand_landmarks in results.multi_hand_landmarks:
                # Draw landmarks
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                
                # Get bounding box of hand
                x_min, y_min = frame.shape[1], frame.shape[0]
                x_max, y_max = 0, 0
                
                for landmark in hand_landmarks.landmark:
                    x, y = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])
                    x_min = min(x_min, x)
                    y_min = min(y_min, y)
                    x_max = max(x_max, x)
                    y_max = max(y_max, y)
                
                # Add padding to bounding box
                padding = 30
                x_min = max(0, x_min - padding)
                y_min = max(0, y_min - padding)
                x_max = min(frame.shape[1], x_max + padding)
                y_max = min(frame.shape[0], y_max + padding)
                
                # Draw bounding box
                cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
                
                # Crop hand region
                hand_crop = frame[y_min:y_max, x_min:x_max]
                
                if hand_crop.size > 0:  # Ensure crop is valid
                    # Convert to grayscale
                    hand_gray = cv2.cvtColor(hand_crop, cv2.COLOR_BGR2GRAY)
                    
                    # Preprocess for prediction
                    processed_data = preprocess_image(hand_gray, img_size, min_value)
                    
                    # Make prediction
                    prediction = model.predict(processed_data)[0]
                    confidence = np.max(prediction)
                    label_index = np.argmax(prediction)
                    
                    # Only consider high confidence predictions
                    if confidence > confidence_threshold:
                        predicted_letter = labels_dict[label_index]
                        recent_predictions.append(label_index)
                        
                        # Get most common prediction from recent frames
                        if recent_predictions:
                            smoothed_label_index = Counter(recent_predictions).most_common(1)[0][0]
                            smoothed_letter = labels_dict[smoothed_label_index]
                            
                            # Display the predicted letter and confidence
                            text = f"{smoothed_letter} ({confidence:.2f})"
                            cv2.putText(frame, text, (x_min, y_min - 10), 
                                         cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                            
                            # Track letter for sentence building
                            if smoothed_letter == prev_letter:
                                letter_counter += 1
                                if letter_counter >= letter_threshold and prev_letter != sentence[-1:]:
                                    if smoothed_letter == '0':  # Use 0 as backspace
                                        if sentence:
                                            sentence = sentence[:-1]
                                    else:
                                        sentence += smoothed_letter
                                    letter_counter = 0  # Reset counter
                            else:
                                prev_letter = smoothed_letter
                                letter_counter = 0
                    else:
                        cv2.putText(frame, f"Low confidence", (x_min, y_min - 10),
                                     cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        else:
            # No hand detected
            space_counter += 1
            if space_counter >= space_threshold and sentence and sentence[-1] != ' ':
                sentence += ' '
                space_counter = 0  # Reset counter

                words = sentence.strip().split(' ')
                if len(words) > 0:
                    last_word = words[-1]
                    if last_word and last_word != last_spoken_text:
                         tts_engine.say(last_word)
                         tts_engine.runAndWait()
                         last_spoken_text = last_word
        
        # Display the sentence at the bottom of the frame
        cv2.rectangle(frame, (0, frame.shape[0] - 50), (frame.shape[1], frame.shape[0]), (0, 0, 0), -1)
        cv2.putText(frame, sentence, (10, frame.shape[0] - 20),
                     cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
        # Add this instruction text
        cv2.putText(frame, "Press 'S' to speak full sentence", (frame.shape[1] - 350, frame.shape[0] - 20),
                     cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200, 200, 200), 1)
        # Show the original frame and processed frame
        cv2.imshow('Sign Language Detection', frame)
        
        # Exit on 'q' press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        elif cv2.waitKey(1) & 0xFF == ord('s'):  # 's' key to speak the entire sentence
            if sentence.strip() != last_spoken_text:
                 tts_engine.say(sentence)
                 tts_engine.runAndWait()
                 last_spoken_text = sentence.strip()
    
    # At the end of the function before closing, add this to release resources
    tts_engine.stop()
    
    cap.release()
    cv2.destroyAllWindows()
    return sentence


```

```python
# To run the real-time detection:
real_time_sign_detection()
```

```python
# PART 5: REPROCESSING THE DATA AND TRAINING THE MODEL WITH THE IMPROVEMENTS
def reprocess_dataset_and_train():
    """Function to reprocess your existing dataset with consistent preprocessing"""
    import os
    import numpy as np
    import cv2
    
    # Paths
    data_path = r"C:\Users\User\minor_project\newdataset"
    categories = os.listdir(data_path)
    
    # Parameters
    img_size = 128
    min_value = 70
    
    # Lists to store data
    data = []
    target = []
    
    # Process each category
    for category in categories:
        cat_path = os.path.join(data_path, category)
        cat_names = os.listdir(cat_path)
        
        for classes in cat_names:
            class_index = int(classes) if classes.isdigit() else ord(classes) - ord('A') + 1
            folder_path = os.path.join(data_path, category, classes)
            img_names = os.listdir(folder_path)
            
            for img_name in img_names:
                img_path = os.path.join(folder_path, img_name)
                img = cv2.imread(img_path)
                
                try:
                    # Use consistent preprocessing function
                    processed = preprocess_image(img, img_size, min_value)
                    data.append(processed[0])  # Remove the batch dimension
                    target.append(class_index)
                except Exception as e:
                    print(f'Exception processing {img_path}: {e}')
    
    # Convert to numpy arrays
    data = np.array(data)
    target = np.array(target)
    
    # Convert targets to one-hot encoding
    from tensorflow.keras.utils import to_categorical
    target_one_hot = to_categorical(target)
    
    # Save processed data
    np.save('data_img_improved.npy', data)
    np.save('target_improved.npy', target_one_hot)
    
    print(f"Dataset processed. Data shape: {data.shape}, Target shape: {target_one_hot.shape}")
    
    # Train model with augmentation
    model, history = train_model_with_augmentation(data, target_one_hot)
    
    # Plot training results
    import matplotlib.pyplot as plt
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('improved_model_performance.png')
    plt.show()
    
    return model, history

```
